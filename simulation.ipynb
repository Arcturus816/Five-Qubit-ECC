```python
probs = [0.025 * x for x in range(1,13)]
probs
```

```python
# Takes 80 s to run 1000 simulations
# 250 simulations per probability for 12 probabilities should take 4 min
more_results = five_qubit_ecc.simulation_results(probs=probs, num_sims=250)
```

```python
plt.plot(*zip(*sorted(more_results.items())))
```

```python
plt.scatter(*zip(*sorted(more_results.items())))
```

```python
plt.plot(*zip(*sorted(more_results.items())))
plt.xlabel("Error probability for each physical qubit")
plt.ylabel("Probability of measuring to a correct component")
plt.title("Simulation results")
plt.grid(True)
plt.show()
```

```python
results_alt = dict()
for key in more_results:
    results_alt.update({1-3*key: more_results.get(key)})

plt.plot(*zip(*sorted(results_alt.items())))
plt.xlabel("1-3p")
plt.ylabel("Probability of measuring to a correct component")
plt.title("Simulation results")
plt.grid(True)
plt.show()
```
